##Eliminando las columnas innecesarias

```python
df_final = datos.drop(columns=['customerID'])
```

```python
categorias = df_final.select_dtypes(include=['object']).columns
```

```python
pd.get_dummies(data=df_final, columns=categorias, dtype = int).head()
```

```python
datos_codificados = pd.get_dummies(data=df_final, columns=categorias, dtype = int)
datos_codificados.sample(5)
```

```python
datos_codificados.info()
```

##Realizamos el encoding

```python
from sklearn.dummy import DummyRegressor

from sklearn.model_selection import train_test_split

X = datos_codificados.drop('Churn', axis=1)
y=datos_codificados['Churn']


X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)

baseline=DummyRegressor()
baseline.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

y_pred_dummy = baseline.predict(X_test)

def calcular_regresion(y_test,y_pred):
  rmse=mean_squared_error(y_test,y_pred)
  mae=mean_absolute_error(y_test,y_pred)
  r2= r2_score(y_test, y_pred)
  metricas = {
      'RMSE':round(rmse**(1/2),4),
      'MAE': round(mae,4),
      'R2':round(r2,4)
  }
  return metricas
```

```python
resultados_baseline=calcular_regresion(y_test,y_pred_dummy)
```

```python
resultados_baseline
```

```python
from sklearn.ensemble import RandomForestRegressor

modelo = RandomForestRegressor(max_depth=5,random_state=42)

modelo.fit(X_train,y_train)

y_pred = modelo.predict(X_test)
resultados_rf=calcular_regresion(y_test,y_pred)
resultados_rf
```

Balanceo de Clases con SMOTE y Over_sampling

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

smote = SMOTE(random_state=42)

X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

print(f"Distribución original: {Counter(y_train)}")
print(f"Distribución balanceada: {Counter(y_train_res)}")
```

##Normalización o Estandarización

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
scaler = StandardScaler()

X_train_res_scaled = scaler.fit_transform(X_train_res)
X_test_scaled = scaler.transform(X_test)

lr_final = LogisticRegression(random_state=42)
lr_final.fit(X_train_res_scaled, y_train_res)

y_pred_lr_final = lr_final.predict(X_test_scaled)

print("--- Evaluación Final: LogReg + SMOTE + Scaler ---")
print(classification_report(y_test, y_pred_lr_final))
```

## Uso del modelo de Regresión Logística

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

modelo_logistico = LogisticRegression(random_state=42)
modelo_logistico.fit(X_train_res_scaled, y_train_res)

y_pred_log = modelo_logistico.predict(X_test_scaled)
```

```python

print("--- Informe de Clasificación: Regresión Logística ---")
print(classification_report(y_test, y_pred_log))

sns.heatmap(confusion_matrix(y_test, y_pred_log), annot=True, fmt='d', cmap='Blues')
plt.title('Matriz de Confusión - Regresión Logística')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

modelo_rf = RandomForestClassifier(n_estimators=100, random_state=42)

modelo_rf.fit(X_train_res_scaled, y_train_res)

y_pred_rf = modelo_rf.predict(X_test_scaled)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print("--- Informe de Clasificación: Random Forest ---")
print(classification_report(y_test, y_pred_rf))

sns.heatmap(confusion_matrix(y_test, y_pred_rf), annot=True, fmt='d', cmap='Greens')
plt.title('Matriz de Confusión - Random Forest')
plt.show()
```
